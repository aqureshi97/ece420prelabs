```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write

FRAME_SIZE = 2048
E_base = 100000000
def EnergyPeak(frame):
     return np.sum(frame*frame)

def ece420ProcessFrame(frame):
    isVoiced = 0
    Amplitude = EnergyPeak(frame)
    if (Amplitude > E_base):
        isVoiced = 1
        return Amplitude > E_base
#### YOUR CODE HERE ####

################# GIVEN CODE BELOW #####################

Fs, data = read('test_vector.wav')

numFrames = int(len(data) / FRAME_SIZE)
framesVoiced = np.zeros(numFrames)

for i in range(numFrames):
    frame = data[i * FRAME_SIZE : (i + 1) * FRAME_SIZE]
    framesVoiced[i] = ece420ProcessFrame(frame.astype(float))

plt.figure()
plt.stem(framesVoiced)
plt.show()
```
