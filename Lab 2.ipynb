```python
#second assignment
# My filter design incroporates an order of 10 and is able to reduce the amplitude of the signal of the humming from 180 dB. 
# I chose 10 taps because the more taps you ahve the better the filtering that occurs, however, it takes much longer as you increase the 
# amount of taps that are present in the filter. This means the cost of the filter will increase. 
# I can't achieve the same effect with less taps.
from scipy import signal
import matplotlib.pyplot as plt
Ny = sr*.5
Fstart = 310 / Ny
Fstop = 490 / Ny
b,a = signal.butter(15,[Fstart, Fstop],'bandstop', analog = True)
w,h = signal.freqs(b,a)
plt.subplot(211)
plt.title("Magnitude response")
plt.xlabel("Radians per second)")
plt.ylabel("Decibels")
plt.plot(w,20*numpy.log10(h))
plt.subplot(212)
plt.title("phase response")
plt.xlabel("frequency(radians per second)")
plt.ylabel("phase angle")
plt.plot(w, numpy.angle(h.imag/h.real))
```

```python
#Second Assignment
# My filter design incroporates an order of 10 and is able to reduce the amplitude of the signal of the humming from 180 dB. #
from scipy.io.wavfile import read
sr, data = read('with_hum.wav')
import numpy 
import scipy
import matplotlib.pyplot as plt
Magnitude = numpy.fft.fft(data)
Magnitude = numpy.fft.fftshift(Magnitude)
l = len(Magnitude)
x = numpy.linspace(-l/2,l/2,l)
Decibels = 20 * numpy.log10(Magnitude)
from scipy import signal
import matplotlib.pyplot as plt
Ny = sr*.5
Fstart = 310 / Ny
Fstop = 490 / Ny
b,a = signal.butter(10,[Fstart, Fstop],'bandstop', analog = True)
w,h = signal.freqs(b,a)
plt.subplot(211)
plt.plot(w,20*numpy.log10(h))
plt.subplot(212)
plt.plot(w, numpy.angle(h.imag/h.real))
Output = signal.filtfilt(b,a,Decibels)
Foutput = signal.lfilter(b,a,Output)
plt.plot()
```
